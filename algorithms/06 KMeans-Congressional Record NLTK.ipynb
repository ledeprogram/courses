# Processing text

If we want to use k-means on text, there are about ten thousand ways to do it.

The first way we'll work with it is to put it into a TfidfVectorizer.

Ttf-idf means **term frequency–inverse document frequency**. The more a term appears in a document, the more important it is. The more is appears in the *total corpus*, the less important it is.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# max_features takes the top x words to process. Useful if you have a huge corpus.
vectorizer = TfidfVectorizer(max_features=10000, stop_words='english', use_idf=True)
```

```python
X = vectorizer.fit_transform(speeches_df['contents'])
# Why do they always use X? They always do.
```

```python
print type(X)
print type(X[0])
# What's a sparse matrix?
```

```python
X[0].data
```

Whatever this sparse matrix is, we can fit it into a KMeans clusterer.

```python
from sklearn.cluster import KMeans

number_of_clusters = 2

km = KMeans(n_clusters=number_of_clusters)
km.fit(X)
```

```python
print("Top terms per cluster:")
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(number_of_clusters):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print ''
```

# Tokenizing and stemming

Let's start simple, by working on a single speech. The fourth one seems plenty long to work with.

```python
speeches_df.ix[3]["contents"]
```

Let's import a lot of stuff from NLTK

```python
from nltk import word_tokenize, wordpunct_tokenize
from nltk.stem.snowball import SnowballStemmer 
import nltk
import re

stemmer = SnowballStemmer("english")
stopwords = nltk.corpus.stopwords.words('english')
```

```python
speech = speeches_df.ix[3]
```

word_tokenize tokenizes words. Fun, right? Well...

```python
print word_tokenize(speech['contents'])[:10]
```

It takes punctuation as words, that's no good. Let's strip out anything that isn't a word character.

```python
stripped = re.sub(r"[^\w ]","", speech['contents'])
stripped[:150]
```

and then tokenize that

```python
tokens = word_tokenize(stripped)
print tokens[:20]
```

We can strip out stopwords while we're at it - while the TfidfVectorizer does that automatically, we might want to do more processing on the tokens.

```python
tokens_without_stopwords = [word for word in tokens if word not in stopwords]
print tokens_without_stopwords[:20]
```

Now we're running lean! Let's **stem** them while we're at it using the **SnowballStemmer**. That way *party* and *parties* will end up as the same word.

```python
stemmed = [stemmer.stem(word) for word in tokens_without_stopwords]
print stemmed[:20]
```

Let's write a function that takes a dataframe and tokenizes + stems it

```python
def speech_to_tokens(speech):
    stripped = re.sub(r"[^\w ]","", speech['contents'])
    tokens = word_tokenize(stripped)
    tokens_without_stopwords = [word for word in tokens if word not in stopwords]
    return [stemmer.stem(word) for word in tokens_without_stopwords]
```

```python
tokens = speech_to_tokens(speeches_df.ix[0])
print tokens[:10]
tokens = speech_to_tokens(speeches_df.ix[1])
print tokens[:10]
tokens = speech_to_tokens(speeches_df.ix[2])
print tokens[:10]
```

But how do we get it to apply to every single row of the dataframe? You actually can't just use a for loop! Remember how you need to use `iterrows()` to loop through a dataframe? That also prevents you from saving data back to it. What you'll need to do is use **apply**.

```python
# Store tokens into the dataframe, applying speech_to_tokens function
# to each row in the dataframe
speeches_df["tokens"] = speeches_df.apply(speech_to_tokens, axis=1)
```

```python
speeches_df[["contents","tokens"]][:10]
```

Beautiful!

# Poking around

Counter can be used to find the most common words. You'll see FreqDist used for the same purpose in some NLTK writeups, but they're working on replacing it with Counter these days.

Let's find the most common 10 tokens in the very first speech.

```python
from collections import Counter
tokens = speech_to_tokens(speeches_df.ix[0])
count = Counter(tokens)
count.most_common(10)
```

But what if we wanted to find it across **all** of the speeches? Let's write a function to do that.

```python
def flatten(dataframe):
    # I still can't wrap my head around how this works
    return [y for x in dataframe for y in x]

flatten(speeches_df['tokens'])[:10]
```

```python
all_tokens = flatten(speeches_df['tokens'])
count = Counter(all_tokens)
count.most_common(30)
```

Now if we really wanted to, we could change the list of stopwords based on this. If we want to be thorough we'll want to stem the stopwords.

```python
original_stopwords = [stemmer.stem(word) for word in nltk.corpus.stopwords.words('english')]
print original_stopwords
```

```python
added_stopwords = ['mr', 'bill', 'gentleman', 'gentlewoman', 'yield', 'speaker', 'bill', 'would', 'chairman', 'amend']
new_stopwords = added_stopwords + original_stopwords
```

# Back to K-Means

```python
# TfidfVectorizer can be passed 'english' for stop_words, or it can be passed
# an array
vectorizer = TfidfVectorizer(max_features=10000, stop_words=new_stopwords, use_idf=True)
X = vectorizer.fit_transform(speeches_df['contents'])

number_of_clusters = 5

km = KMeans(n_clusters=number_of_clusters)
km.fit(X)

print("Top terms per cluster:")
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(number_of_clusters):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print ''
```

# Decision Trees

Idea: Maybe it's because there are so many reallllly short speeches. Maybe we need a **decision tree**. Smaller word counts are kicked to the curb while we keep the longer, meatier ones.

```python
number_of_clusters = 10
X = vectorizer.fit_transform(speeches_df[speeches_df['word_count'] > 100]['contents'])
km = KMeans(n_clusters=number_of_clusters)
km.fit(X)
print("Top terms per cluster:")
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(number_of_clusters):
    print("Cluster %d:" % i),
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind]),
    print ''
```
